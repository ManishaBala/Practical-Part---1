###2.Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import randint

x = np.arange(1, 7)  # Outcomes from 1 to 6 (like a dice roll)
pmf = randint.pmf(x, 1, 7)  # Discrete Uniform PMF from 1 to 6

plt.stem(x, pmf)  # <- Fixed line
plt.title("PMF of Discrete Uniform Distribution (1-6)")
plt.xlabel("x")
plt.ylabel("PMF")
plt.show()
```

###3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
from scipy.stats import bernoulli

def bernoulli_pdf(p):
    x = [0, 1]
    return x, bernoulli.pmf(x, p)

x, pdf = bernoulli_pdf(0.6)
print("x:", x)
print("PDF:", pdf)
```

###9. Write a Python function to calculate Z-scores from a dataset and plot them.

```python
def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    return [(x - mean) / std for x in data]

z_scores = calculate_z_scores(data)
plt.hist(z_scores, bins=30, edgecolor='black')
plt.title("Histogram of Z-Scores")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()
```

###10.  Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

```python
means = []
for _ in range(1000):
    sample = np.random.exponential(scale=2, size=30)
    means.append(np.mean(sample))

plt.hist(means, bins=30, edgecolor='black')
plt.title("Central Limit Theorem (Means of Exponential Samples)")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```

###15. Implement hypothesis testing using Z-statistics for a sample dataset.

```python
from scipy.stats import norm

sample_mean = 105
population_mean = 100
std = 15
n = 30
z = (sample_mean - population_mean) / (std / np.sqrt(n))
p_value = 1 - norm.cdf(z)

print("Z-score:", z)
print("P-value:", p_value)
```

###16. Create a confidence interval for a dataset using Python and interpret the result.

```python
import scipy.stats as stats

data = np.random.normal(100, 15, 50)
mean = np.mean(data)
se = stats.sem(data)
ci = stats.t.interval(0.95, len(data)-1, loc=mean, scale=se)

print("95% Confidence Interval:", ci)
```

###17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

```python
import numpy as np
from scipy import stats

# Step 1: Generate data from a normal distribution
data = np.random.normal(loc=100, scale=15, size=50)  # mean=100, std=15, n=50

# Step 2: Calculate sample mean
sample_mean = np.mean(data)

# Step 3: Calculate standard error of the mean
standard_error = stats.sem(data)

# Step 4: Calculate 95% confidence interval
confidence_level = 0.95
degrees_freedom = len(data) - 1

confidence_interval = stats.t.interval(confidence_level, df=degrees_freedom,
                                       loc=sample_mean, scale=standard_error)

# Output the result
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval for the Mean: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})")
```

###18. Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Step 1: Generate normally distributed data
data = np.random.normal(loc=0, scale=1, size=1000)  # mean = 0, std = 1

# Step 2: Create x-values for the PDF curve
x = np.linspace(-4, 4, 1000)

# Step 3: Calculate the PDF
pdf = norm.pdf(x, loc=np.mean(data), scale=np.std(data))

# Step 4: Plot the histogram and PDF
plt.hist(data, bins=30, density=True, alpha=0.6, color='skyblue', label='Histogram')
plt.plot(x, pdf, 'r-', lw=2, label='PDF')
plt.title("Normal Distribution with PDF")
plt.xlabel("x")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```
